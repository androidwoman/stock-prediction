# gpt facts of train baba data to csv

```python
import re
import csv

# Function to parse the input text and convert it to a structured format
def parse_input(text):
    facts = []
    entries = re.split(r'### Fact \d+:', text)
    for entry in entries[1:]:  # Skip the first split as it will be empty
        summary_match = re.search(r'\*\*Summary:\*\* ([\s\S]*?)\n\n', entry)
        price_movement_match = re.search(r'\*\*Price Movement:\*\* (\w+)\n\n', entry)
        explanation_match = re.search(r'\*\*Explanation:\*\* ([\s\S]*?)(?=\n### Fact \d+:|$)', entry)
        
        summary = summary_match.group(1).replace('\n', ' ').strip() if summary_match else ''
        price_movement = price_movement_match.group(1).strip() if price_movement_match else ''
        explanation = explanation_match.group(1).replace('\n', ' ').strip() if explanation_match else ''
        
        facts.append({
            'Summary': summary,
            'Price Movement': price_movement,
            'Explanation': explanation
        })
    return facts

# Read the input text from the file
with open('input.txt', 'r') as file:
    input_text = file.read()

# Parse the input text
parsed_data = parse_input(input_text)

# Write the parsed data to a CSV file
with open('output.csv', 'w', newline='') as csvfile:
    fieldnames = ['Summary', 'Price Movement', 'Explanation']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for fact in parsed_data:
        writer.writerow(fact)

print("Data successfully written to output.csv")
```

# gpt tweet to movement and exp

```python
import re
import pandas as pd

def parse_fact_block(block):
    fact_match = re.search(r'\*\*Fact:\s*"(.*?)"\s*\*\*', block, re.DOTALL)
    start_date_match = re.search(r'\*\*Start Date:\s*(.*?)\*\*', block)
    end_date_match = re.search(r'\*\*End Date:\s*(.*?)\*\*', block)
    price_movement_match = re.search(r'\*\*Price Movement:\s*(.*?)\*\*', block)
    explanation_match = re.search(r'\*\*Explanation:\s*(.*?)\*\*', block, re.DOTALL)
    confidence_match = re.search(r'Confidence:\s*(\d+\.?\d*)%', block)

    fact = fact_match.group(1).strip() if fact_match else ""
    start_date = start_date_match.group(1).strip() if start_date_match else ""
    end_date = end_date_match.group(1).strip() if end_date_match else ""
    price_movement = price_movement_match.group(1).strip() if price_movement_match else ""
    explanation = explanation_match.group(1).strip() if explanation_match else ""
    confidence = confidence_match.group(1).strip() if confidence_match else ""

    return {
        "Fact": fact,
        "Start Date": start_date,
        "End Date": end_date,
        "Price Movement": price_movement,
        "Explanation": explanation,
        "Confidence": confidence
    }

# Read the text file
with open("100_first_gpt.txt", "r") as file:
    content = file.read()

# Use regex to split the content into blocks based on the pattern
blocks = re.split(r'\n\s*\d+\.\s\*\*Fact:', content)
blocks = ["**Fact:" + block.strip() for block in blocks if block.strip()]

data = [parse_fact_block(block) for block in blocks]

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
csv_path = "100_first_gpt.csv"
df.to_csv(csv_path, index=False)

print(f"Data has been saved to {csv_path}")
```

```python
import pandas as pd

# Read the CSV files
target_file = 'BABA(3).csv'
price_movement_file = '100_first_gpt.csv'


price_movement_df = pd.read_csv(price_movement_file)
target_df = pd.read_csv(target_file)

# Convert date columns to datetime format
price_movement_df['Start Date'] = pd.to_datetime(price_movement_df['Start Date'], errors='coerce')
price_movement_df['End Date'] = pd.to_datetime(price_movement_df['End Date'], errors='coerce')
target_df['start'] = pd.to_datetime(target_df['start'], errors='coerce')
target_df['end'] = pd.to_datetime(target_df['end'], errors='coerce')

# Ensure there are only 100 items in the target DataFrame
target_df = target_df.head(100)

# Merge the DataFrames on 'Start Date' and 'End Date' as keys
merged_df = pd.merge(price_movement_df, target_df, left_on=['Start Date', 'End Date'], right_on=['start', 'end'], how='inner')



```

```python
# Function to check price movement
def check_price_movement(row):
    return row['Price Movement'] == row['target']

# Apply the function to each row
merged_df['Check'] = merged_df.apply(check_price_movement, axis=1)

# Count the number of correct and incorrect predictions
correct_count = merged_df['Check'].sum()
incorrect_count = len(merged_df) - correct_count

# Print the counts
print(f"Correct predictions: {correct_count}")
print(f"Incorrect predictions: {incorrect_count}")



# Save the result to a new CSV file
merged_df.to_csv('price_movement_check.csv', index=False)
```

```python
import matplotlib as plt

# Aggregate counts by Confidence and Check
confidence_agg = merged_df.groupby(['Confidence', 'Check']).size().unstack(fill_value=0)

# Plotting
confidence_agg.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.title('Correct and Incorrect Predictions by Confidence')
plt.xlabel('Confidence')
plt.ylabel('Number of Predictions')
plt.legend(['Incorrect', 'Correct'])
plt.show()
```

```python

# Calculate the percentage of positive and negative targets
target_counts = merged_df['target'].value_counts(normalize=True) * 100

# Count the number of correct and incorrect predictions for positive and negative targets
positive_correct = merged_df[(merged_df['target'] == 'Positive') & (merged_df['Check'] == True)].shape[0]
positive_incorrect = merged_df[(merged_df['target'] == 'Positive') & (merged_df['Check'] == False)].shape[0]
negative_correct = merged_df[(merged_df['target'] == 'Negative') & (merged_df['Check'] == True)].shape[0]
negative_incorrect = merged_df[(merged_df['target'] == 'Negative') & (merged_df['Check'] == False)].shape[0]

# Plot the percentage of positive and negative targets
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
target_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Percentage of Positive and Negative Targets')
plt.ylabel('')

# Plot the count of correct and incorrect predictions for positive and negative targets
correct_incorrect_counts = pd.DataFrame({
    'Correct': [positive_correct, negative_correct],
    'Incorrect': [positive_incorrect, negative_incorrect]
}, index=['Positive', 'Negative'])

plt.subplot(1, 2, 2)
correct_incorrect_counts.plot(kind='bar', stacked=True, color=['green', 'red'], ax=plt.gca())
plt.title('Correct and Incorrect Predictions by Target')
plt.xlabel('Target')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Ensure there are only 100 items in the target DataFrame
target_df = target_df.head(100)

# Merge the DataFrames on 'Start Date' and 'End Date' as keys
merged_df = pd.merge(price_movement_df, target_df, left_on=['Start Date', 'End Date'], right_on=['start', 'end'], how='inner')

# Convert Confidence to numeric
merged_df['Confidence'] = pd.to_numeric(merged_df['Confidence'], errors='coerce')

# Extract the month from the Start Date
merged_df['Month'] = merged_df['Start Date'].dt.to_period('M')

# Function to check price movement
def check_price_movement(row):
    return row['Price Movement'] == row['target']

# Apply the function to each row
merged_df['Check'] = merged_df.apply(check_price_movement, axis=1)

# Count the percentage of positive and negative targets by month
monthly_target_counts = merged_df.groupby(['Month', 'target']).size().unstack(fill_value=0)
monthly_target_percentages = monthly_target_counts.div(monthly_target_counts.sum(axis=1), axis=0) * 100

# Count the number of correct and incorrect predictions by month and target
monthly_correct_counts = merged_df.groupby(['Month', 'target', 'Check']).size().unstack(fill_value=0).fillna(0)
monthly_correct_counts.columns = ['Incorrect', 'Correct']

# Plotting the percentage of positive and negative targets by month
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
monthly_target_percentages.plot(kind='bar', stacked=True, color=['lightcoral', 'skyblue'], ax=plt.gca())
plt.title('Percentage of Positive and Negative Targets by Month')
plt.xlabel('Month')
plt.ylabel('Percentage')
plt.legend(title='Target')

# Plotting the count of correct and incorrect predictions by month and target
plt.subplot(1, 2, 2)
monthly_correct_counts.plot(kind='bar', stacked=True, color=['red', 'green'], ax=plt.gca())
plt.title('Correct and Incorrect Predictions by Month and Target')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(title='Prediction')

plt.tight_layout()
plt.show()
```
