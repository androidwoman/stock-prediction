```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.plot(x,y,'o')
plt.show()
```

```python
plt.plot(y,'o')
plt.show()
```

```python
plt.plot(y,marker='o',linestyle='--',color='g',label='y')
plt.show()
```

```python
plt.plot(y,'o:r')
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Two lines on same plot with legend')
plt.grid(axis='y',color='r',linestyle='--',linewidth=0.5)
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.subplot(1, 2, 1)
plt.plot(x1, y1, label = "line 1")
plt.subplot(1, 2, 2)
plt.plot(x2, y2, label = "line 2")
plt.show()
```

```python
colors=[0,1,2,3]
plt.scatter(x, y, cmap='viridis',c=colors)
plt.colorbar()
plt.show()
```

```python
x=np.array(['first','second','third','fourth'])
y=np.array([1,2,3,4])
plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
x=np.random.normal(loc=170, scale=10, size=1000)
plt.hist(x)
plt.show()
```

```python
x=np.array(['apple','banana','cherry'])
y=np.array([10,20,70])
plt.pie(y, labels = x,explode=[0,0.1,0],shadow=True)
plt.legend(title = "Fruits")
plt.show()
```
