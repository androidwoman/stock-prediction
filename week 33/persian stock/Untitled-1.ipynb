```python
import pytse_client as tse
```

```python
tickers = tse.download(symbols="all", write_to_csv=True)
```

```python
best_stocks_user_provided = [
    'فولاد',  # فولاد مبارکه اصفهان
    'فملی',  # ملی صنایع مس ایران
    'رمپنا',  # گروه مپنا
    'فارس',  # پتروشیمی خلیج فارس
    'وبملت',  # بانک ملت
    'شپنا',  # پالایش نفت اصفهان
    'خساپا',  # سایپا
    'خودرو',  # ایران خودرو
    'میدکو',  # هلدینگ صنایع معدنی خاورمیانه
    'ومعادن'  # توسعه معادن و فلزات
]
```

```python
import os
folder_name = "10_selected"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
```

```python
for stock in best_stocks_user_provided:
    if stock in tickers:
        stock_data = tickers[stock]
        stock_data.to_csv(f"{folder_name}/{stock}.csv", index=False)
```
